```python

import numpy as np
import matplotlib.pyplot as plt
from utils import SimulationAgent, execute_multiple_trials, compute_average_curve, execute_cross_entropy_trials

# Experiment code

# Run trials with different covariance values
experiment_results = []
for i in range(1, 11):
    test_covariance = i / 10
    gains, parameters = execute_multiple_trials(test_covariance)
    experiment_results.append((gains, parameters))
    print(f"execute_multiple_trials done for Sigma = {test_covariance}")
```

```python

# Plot the results with different styles and layouts
fig, axs = plt.subplots(5, 2, figsize=(12, 14), constrained_layout=True)
axs = axs.flatten()
for i in range(10):
    gains = experiment_results[i][0]
    x = [gains[j][0] for j in range(len(gains))]
    y = [gains[j][1] for j in range(len(gains))]
    axs[i].scatter(x, y, label=f"Sigma: {i / 10:.1f}", color=plt.cm.viridis(i / 10))
    axs[i].legend(loc='upper left')
    axs[i].set_xlabel('Trial #')
    axs[i].set_ylabel('Expected Gain')
plt.show()
```

```python

# Get averaged learning curve with different plotting style
gains, parameters = execute_multiple_trials(0.9)
x = [gains[j][0] for j in range(len(gains))]
y = [gains[j][1] for j in range(len(gains))]
plt.plot(x, y, linestyle='--', color='purple', marker='o', label='Expected Gain vs. Trial')
plt.xlabel('Trial #')
plt.ylabel('Expected Gain')
plt.title('Learning Curve for Expected Gain vs. Trial')
plt.legend(loc='upper right')
plt.grid(True, linestyle=':')
plt.show()
```

```python


# Plot average learning curve with shaded error bars
average_learning_curve = compute_average_curve(parameters, 0.9, 300, 30)
num_trials = 300
standard_error = np.std(average_learning_curve) / np.sqrt(num_trials)
plt.plot([i for i in range(1, 31)], average_learning_curve, linestyle='-', color='darkred', label='Expected Gain vs. Episode')
plt.fill_between([i for i in range(1, 31)], 
                 np.array(average_learning_curve) - standard_error, 
                 np.array(average_learning_curve) + standard_error, 
                 color='darkred', alpha=0.2)
plt.xlabel('Episode #')
plt.ylabel('Expected Gain')
plt.title('Expected Gain vs. Episode with Shaded Error')
plt.legend(loc='lower right')
plt.grid(True, linestyle='--')
plt.show()
```

```python

# Cross-entropy method with different plotting style
new_gains, new_params, new_cov = execute_cross_entropy_trials(0.99, 10)
x = [new_gains[j][1] for j in range(len(new_gains))]
y = [new_gains[j][0] for j in range(len(new_gains))]
plt.plot(x, y, linestyle=':', color='green', marker='x', label='Expected Gain vs. Trial')
plt.xlabel('Trial #')
plt.ylabel('Expected Gain')
plt.title('Learning Curve for Expected Gain vs. Trial (Cross-Entropy Method)')
plt.legend(loc='lower left')
plt.grid(True, linestyle='-.')
plt.show()
```

```python

# Plot average learning curve for cross-entropy method with error bars
average_learning_curve = compute_average_curve(new_params, new_cov, 300, 30)
num_trials = 300
standard_error = np.std(average_learning_curve) / np.sqrt(num_trials)
plt.plot([i for i in range(1, 31)], average_learning_curve, linestyle='-', color='darkblue', label='Expected Gain vs. Episode')
plt.errorbar([i for i in range(1, 31)], average_learning_curve, yerr=standard_error, ecolor='orange', fmt='o', alpha=0.6)
plt.xlabel('Episode #')
plt.ylabel('Expected Gain')
plt.title('Expected Gain vs. Episode with Error Bars')
plt.legend(loc='upper left')
plt.grid(True, linestyle='-')
plt.show()
```
